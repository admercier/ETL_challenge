<h3>Second DataSource (CSV)

```python
#read in csv file
florida_amps = pd.read_csv(r"C:\Users\Abbey Mercier\Documents\GitHub\ETL_Project\LE Amputation Diabetes Florida - County.csv")
florida_amps
```

```python
#drop uwanted columns
florida_amps = florida_amps.drop(columns=["Unnamed: 1","Unnamed: 2", "Unnamed: 9", "Unnamed: 10",
                          "Unnamed: 11", "Unnamed: 12", "Unnamed: 13", "Unnamed: 14", "Unnamed: 15",
                          "Unnamed: 16", "Unnamed: 17", "Unnamed: 18", "Unnamed: 19", "Unnamed: 20"])
florida_amps
```

```python
#drop unwanted rows
florida_amps = florida_amps.drop([0,1])
florida_amps
```

```python
#rename columns
florida_amps = florida_amps.rename(columns={"Age-adjusted Hospitalizations From Amputation of a Lower Extremity Attributable to Diabetes, Rate Per 100,000 Population":"County",
                                            "Unnamed: 3":"2018 Count","Unnamed: 4": "2018 Rate",
                                            "Unnamed: 5": "2017 Count", "Unnamed: 6": "2017 Rate", 
                                            "Unnamed: 7": "2016 Count", "Unnamed: 8": "2016 Rate"})
florida_amps
```

```python
#confirm no rows need to be removed
florida_amps.info()
```

<h3> Join Two Data Sets

```python
#join two dataframes together into on dataframe
florida_join = pd.merge(florida_income, florida_amps, how="left", on="County")
florida_join

#reindex to just the county
florida_join.set_index("County", inplace=True)
florida_join
```

```python
#save dataframe as CSV
florida_join.to_csv("Florida_Amputation_and_Income.csv")
```
